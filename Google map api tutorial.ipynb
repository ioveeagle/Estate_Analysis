```python
import requests
import urllib
import json
import requests.packages.urllib3
requests.packages.urllib3.disable_warnings()
```

```python
def get_geographic_coordinates(address):
    #將中文轉成url編碼
    address = urllib.parse.quote(address)
    #my_key = "AIzaSyA9B2j7SnTls-GhyCKGAgtWD2-jMfVDBF8"
    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address}"
    #print(url)
    try:
      res = requests.get(url,verify=False)  
      if res.status_code ==200:
        js = json.loads(res.text)
        #print(js)
        result = js["results"][0]["geometry"]["location"]
        lat = result["lat"]
        #print("緯度:",lat)
        lng = result["lng"]
        #print("經度:",lng)
        return(lat,lng)
    except:
      #print ERROR ADDRESS if address can't be parsed  
      print("ERROR ADDRESS")
```

```python
def get_distance(address1,address2):
    my_key = "AIzaSyA9B2j7SnTls-GhyCKGAgtWD2-jMfVDBF8"
    url = f"https://maps.googleapis.com/maps/api/distancematrix/json?origins={address1[0]}%2C{address1[1]}&destinations={address2[0]}%2C{address2[1]}&key={my_key}"
    #print(url)
    try:
      res = requests.get(url,verify=False)  
      if res.status_code ==200:
        js = json.loads(res.text)
        #print(js)
        distance = js["rows"][0]["elements"][0]["distance"]["value"]
        result={"單位:公尺":distance,"單位:10公尺":distance/10,"單位:50公尺":distance/50,"單位:100公尺":distance/100}
        return f"距離:{result}"
    except: 
      print("ERROR")
```

```python
address1=[input("第一個GPS的緯度"),input("第一個GPS的經度")]
address2=[input("第二個GPS的緯度"),input("第二個GPS的經度")]
get_distance(address1,address2)
```
